```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
# Load dataset
df = pd.read_csv('datasheets/diabetes_prediction_dataset.csv')
df
```

```python
df.info()
```

```python
df.head()
```

## Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
gender = encoder.fit_transform(df['gender'])
# Print the encording value
print('Labels : ',encoder.classes_)
print("Encodings : ",encoder.transform(encoder.classes_))
```

```python
smoking_history = encoder.fit_transform(df['smoking_history'])
# Print the encording value
print('Labels : ',encoder.classes_)
print("Encodings : ",encoder.transform(encoder.classes_))
```

```python
# Replace encoded labels with main df
df['gender'] = gender
df['smoking_history'] = smoking_history
```

```python
df.head()
```

# Dataset split

```python
from sklearn.model_selection import train_test_split

y = df['diabetes']
x = df.drop(columns='diabetes')
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=.2,stratify=df['diabetes'])
```

```python
print(x_train.shape)
print(y_train.shape)

print(x_test.shape)
print(y_test.shape)
```

## Feature Scaling (Standardization)

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python

x_train = pd.DataFrame(scaler.fit_transform(x_train),columns=x_train.columns)
x_test = pd.DataFrame(scaler.fit_transform(x_test), columns=x_test.columns)
```

```python
x_train.head()
```

## Decison Tree Classifier

```python
# Step-1 import model
from sklearn.tree import DecisionTreeClassifier

# Step-2: Create an object of the model
model = DecisionTreeClassifier(random_state=42)

# Step-3: fit the model with the training set or train model
model.fit(x_train,y_train)
```

```python
# step-4: get prediction
y_pred = model.predict(x_test)
```

```python
# Step-5: Model performance measure
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred)) # original test result, pred test result
```

```python
# Train and Test Accuracy
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
print("Test Accurecy : ",accuracy_score(y_test,y_pred))
print("Train Accurecy : ",accuracy_score(y_train,model.predict(x_train)))
```

```python
# See confusion Matrix
import scikitplot as skplt
skplt.metrics.plot_confusion_matrix(y_test,y_pred)
```

## Learning Curve

```python
import scikitplot as skplt
plt.figure(figsize=(8,8))

for i in range(1,21):
    clf = DecisionTreeClassifier(random_state=42,max_depth=i)
    text = 'Learning Cureve (Max depth = '+str(i)+')'
    skplt.estimators.plot_learning_curve(clf,x_train,y_train,title=text)
```
