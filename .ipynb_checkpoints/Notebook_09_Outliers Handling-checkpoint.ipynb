## Count Outliers

```python
def count_outliers(data, column):
    column_data = data[column]
    # Calculator  Q1 and Q3
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    # Calculate IQR
    IQR = Q3 - Q1
    # Min and Max Calulation
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # count outliers
    outliers = column_data[(column_data<lower_bound) | (column_data>upper_bound)]
    return len(outliers)
```

```python
attributes = ['age', 'bmi', 'HbA1c_level', 'blood_glucose_level']

for attribute in attributes:
    count = count_outliers(df,attribute)
    print(attribute,'-->',count,'out of',len(df[attribute]))
```

## Remove Outliers

```python
def remove_outliers(data, column):
    column_data = data[column]
    # Calculator  Q1 and Q3
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    # Calculate IQR
    IQR = Q3 - Q1
    # Min and Max Calulation
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    #print(lower_bound,'-------',upper_bound)
    # remove outliers
    cleaned_df = data[(data[column]>= lower_bound) & (data[column] <= upper_bound)]
    return cleaned_df
```

```python
cleaned_df = remove_outliers(df,'bmi')
print("Original df : ",df.shape)
print("Cleaned df : ",cleaned_df.shape)
```

## Before Clean
