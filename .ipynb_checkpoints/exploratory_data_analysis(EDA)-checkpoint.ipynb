## Data Visualilzation

### Diabetes target variable count

```python
labels = ['No Diabetes (0)','Diabetes (1)']
sizes = [df.Diabetes_binary.value_counts()[0],df.Diabetes_binary.value_counts()[1]]

# Create a pie chart
plt.figure(figsize=(10,8))
plt.pie(sizes,labels=labels,autopct='%1.1f%%',startangle=90)
plt.title("Distribution of Diabetes Records")
plt.show()
```

## Diabetes Records per Gender

```python
# 0 = No diabetes
# 1 = diabetes
# 0 = female
# 1 = male

# Create count plot
sns.countplot(data=df,x='Diabetes_binary',hue='Sex',palette='flare')
# Set plot lables
plt.xlabel("Diabetes")
plt.ylabel("Count")
plt.title("Target Variable Count")
plt.legend(title="Gender",labels=["Female","Male"])
plt.xticks(ticks=[0,1],labels=["No Diabetes","Pre-diabetes"])
plt.show()
```

## Relationship between diabetes and stroke

```python
labels = ['No Stroke (0)','Stroke (1)']
sizes = [df.Stroke.value_counts()[0],df.Stroke.value_counts()[1]]

# Create a pie chart
plt.figure(figsize=(10,8))
plt.pie(sizes,labels=labels,autopct='%1.1f%%',startangle=90)
plt.title("Distribution of Stroke Records")
plt.show()
```

```python
# 0 = No diabetes
# 1 = diabetes
# 0 = no stroke
# 1 = stroke

# Create count plot
sns.countplot(data=df,x='Diabetes_binary',hue='Stroke',palette='flare')
# Set plot lables
plt.xlabel("Diabetes Stages")
plt.ylabel("Count")
plt.title("Relationship Between Diabetes and Stroke")
plt.legend(title="Stroke",labels=["NO","YES"])
plt.xticks(ticks=[0,1],labels=["No Diabetes","Pre-Diabetes"])
plt.show()
```

## Relationship between diebetes and highbp

```python
labels = ['No BP (0)','BP (1)']
sizes = [df.HighBP.value_counts()[0],df.HighBP.value_counts()[1]]

# Create a pie chart
plt.figure(figsize=(10,8))
plt.pie(sizes,labels=labels,autopct='%1.1f%%',startangle=90)
plt.title("Distribution of BP Records")
plt.show()
```

```python
# 0 = No diabetes
# 1 = diabetes
# 0 = no bp
# 1 = bp

# Create count plot
sns.countplot(data=df,x='Diabetes_binary',hue='HighBP',palette='flare')
# Set plot lables
plt.xlabel("Diabetes Stages")
plt.ylabel("Count")
plt.title("Relationship Between Diabetes and HighBP")
plt.legend(title="BP",labels=["NO","YES"])
plt.xticks(ticks=[0,1],labels=["No Diabetes","Pre-Diabetes"])
plt.show()
```

## Histogram of BMI


```python
# Create a histogram using seaborn
plt.figure(figsize=(10,8))
sns.histplot(data=df,bins=20,x='BMI',kde=True,color='red') # bins bar size
plt.xlabel("BMI")
plt.title("Histogram")
plt.show()
```
