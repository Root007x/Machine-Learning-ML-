## Which regions have more populations?
##### The mean is the average or the most common value in a collection of numbers. In statistics, it is a measure of central tendency of a probability distribution along median and mode.

```python
# df.groupby(by=grouping_columns)[columns_to_show].function()

df["Population_mln"].mean()

df.groupby(by="Region")["Population_mln"].describe()
```

```python
# groupby agg is a method specifically for performing aggregation(total) operations on a grouped DataFrame.

df.groupby(by="Region")[["Population_mln"]].agg(['mean'])
```

```python
plt.figure(figsize=(8,6))
sns.barplot(data=df,x="Region",y="Population_mln")
plt.title("Region VS Population In Million")
plt.xlabel("Region")
plt.ylabel("Population In Million")
plt.xticks(rotation=45)
plt.show()

```

## What is the average infant deaths across different region?

## How has adult mortality changed over the years?

```python
df.groupby(by="Year")[["Adult_mortality"]].agg(["mean"])
```

```python
# Graph

plt.figure(figsize=(8,6))
sns.lineplot(data=df,x="Year",y="Adult_mortality")
plt.title("Mortality Rate over the years")
plt.xlabel("Years")
plt.ylabel("Adult Mortality")
plt.xticks(rotation=45)
plt.show()
```
