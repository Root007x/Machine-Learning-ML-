# Data Visualization

## Matplotlib

```python
import matplotlib.pyplot as plt

# Set the figure size
plt.figure(figsize=(8,6)) # width,height
# Add labels and title
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Figure Title')
plt.grid()
plt.show()
```

## Bar Charts

```python
import matplotlib.pyplot as plt
import pandas as pd

# Creating a sample dataframe
data = {
    "Categories":['Catagory A','Catagory B','Catagory C','Catagory D'],
    "Values":[25,40,30,35]
}
df = pd.DataFrame(data)

# Plotting the bar chart using matplotlib bar function
plt.bar(df["Categories"],df["Values"])
# Adding labels and title
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Fig : Catagory vs Values")
# Add grid
plt.grid()
# Display the chart
plt.show()
```

```python
# Using Seaborn
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Creating a sample dataframe
data = {
    "Categories":['Catagory A','Catagory B','Catagory C','Catagory D'],
    "Values":[25,40,30,35]
}
df = pd.DataFrame(data)

sns.barplot(x=df.Categories,y=df.Values)
```

## Count Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = {
    "Stroke" : ["Yes","No","Not Sure", "Yes","NO","Not Sure","Yes","No","Maybe","Yes"]
}

df = pd.DataFrame(data)

# Creating a count plot using sns
plt.figure(figsize=(8,6))
sns.countplot(data=df,x=df.Stroke)

plt.title('Stroke Count')
plt.xlabel("Stroke Study")
plt.ylabel("Count")

plt.show()
```

## Hist Plot

```python
import pandas as pd
# import dataset

df = pd.read_csv("healthcare-dataset-stroke-data.csv")

df.info()
```

# Scatter Plot
### visualize the relationship between two continous varaibles

```python
import pandas as pd

df = pd.read_csv("SOCR-HeightWeight.csv")

df.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
sns.scatterplot(data=df,y=df["Height(Inches)"],x=df["Weight(Pounds)"],color="skyblue")
plt.title("Weight vs Height")
plt.xlabel("Weight (Pounds)")
plt.ylabel("Height (Inches)")
plt.show()
```

# Line Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Date" : pd.date_range("2023-01-01",periods=90),
    "Value" : np.random.uniform(low=0,high=100,size=90)
})

# Creating line plot
plt.figure(figsize=(10,6))
sns.lineplot(data=df,x="Date",y="Value",color="blue")
plt.title("Line Plot of Time series Date")
plt.xlabel("Date")
plt.ylabel("Value")

# Rotate xlabel and ylabel
plt.xticks(rotation=45)
plt.yticks(rotation=45)

plt.show()
```

# Swarm plot
### It is categorical scatter plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# import dataset

df = pd.read_csv("healthcare-dataset-stroke-data.csv")

plt.figure(figsize=(10,6))
sns.swarmplot(data=df,x="stroke",y="bmi",palette='Set2',size=1)
plt.title("Swarmplot Stroke vs BMI")
plt.xlabel("Stroke Status")
plt.ylabel("BMI")

plt.show()
```

# BOX Plot

```python
sns.boxplot(x=df.bmi)
plt.xlabel("BMI")
plt.title("Box Plot")
plt.grid()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "A" : np.random.randn(100),
    "B" : np.random.uniform(low=0,high=1,size=100),
    "C" : np.random.normal(loc=2,scale=1,size=100)
})

# Create a pair plot
plt.figure(figsize=(8,6))
sns.pairplot(df)
plt.title("Pair plot of Numerical Variables",y=1.02)
plt.tight_layout()
plt.show()
```
